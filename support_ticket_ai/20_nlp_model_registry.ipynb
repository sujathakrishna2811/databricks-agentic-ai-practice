#### 21_nlp_model_registry

##### Purpose
Register the best Support Ticket NLP classification model from MLflow into Unity Catalog Model Registry.

##### Input
- `support_ticket_ai.best_model_info`
- MLflow artifact path: `support_ticket_model`

##### Output
- Registered model:
  `dbw_agentic_ai_dev.support_ticket_ai.support_ticket_classifier`

##### Architecture

```text

best_model_info
      │
      ▼
best_run_id
      │
      ▼
MLflow Model URI
      │
      ▼
Unity Catalog Model Registry
      │
      ▼
support_ticket_classifier
      │
      ▼
Version 1
```

````python
# import libraries
import mlflow

#Read best_model_info and get best run_id
best_info = spark.table( "dbw_agentic_ai_dev.support_ticket_ai.best_model_info").collect()[0]
best_run_id = best_info["best_run_id"]

# artifact folder
artifact_path="support_ticket_model"

# Construct model URI
model_uri = f"runs:/{best_run_id}/{artifact_path}"

# unity catalog registered model name
registered_model_name = "dbw_agentic_ai_dev.support_ticket_ai.support_ticket_classifier"

# register model
registered_model = mlflow.register_model(
    model_uri=model_uri,
    name=registered_model_name
)

#Print model version
print("Model registered successfully")
print(f"Model URI: {model_uri}")
print(f"Registered model name: {registered_model_name}")
print(f"Model version: {registered_model.version}")
````

##### Notebook Summary

- Read the best model information from Delta.
- Constructed the MLflow model URI.
- Registered the best model into Unity Catalog.
- Created Version 1 of the registered model.
- Model is ready for deployment.

#####  Key Learnings

- Difference between MLflow logged model and Unity Catalog registered model.
- Purpose of artifact_path.
- Purpose of model_uri.
- Model Registry creates versions automatically.
- Registration uses run_id + artifact_path.
