###### Section : Register the MLflow model in Unity Catalog

Why register a model?

Training creates an MLflow logged model tied to a specific experiment run. Registering the model in Unity Catalog creates a governed, versioned model that can be shared, deployed, and managed independently of the training run.

````python
# import libraries
import mlflow

#Read best_model_info and get best run_id
best_info = spark.table( "dbw_agentic_ai_dev.support_ticket_ai.best_model_info").collect()[0]
best_run_id = best_info["best_run_id"]

# artifact folder
artifact_path="support_ticket_model"

# Construct model URI
model_uri = f"runs:/{best_run_id}/{artifact_path}"

# unity catalog registered model name
registered_model_name = "dbw_agentic_ai_dev.support_ticket_ai.support_ticket_classifier"

# Register the MLflow model in Unity Catalog
registered_model = mlflow.register_model(
    model_uri=model_uri,
    name=registered_model_name
)

#Print model version
print("Model registered successfully")
print(f"Model URI: {model_uri}")
print(f"Registered model name: {registered_model_name}")
print(f"Model version: {registered_model.version}")
````

##### Notebook Summary

#####  Key Learnings

- Difference between MLflow logged model and Unity Catalog registered model.
- Artifact_path identifies the logged model inside an MLflow run..
- Model_uri uniquely identifies the logged model to be registered..
- Model Registry creates versions automatically.
- Model registration uses run_id + artifact_path to locate the logged model.
- MLflow Logged Model
        ↓
  Model Registry
        ↓
  Serving Endpoint
