- NLP - Natural Language Processing
- Tokenization --> Split into words. Sentence -> Words- 
- Vocabulary ----> Collect every unique word.- 
- Remove Stop Words ---> Stop words are common words(the, a, is, of, and a, an, to)These usually add little meaning.- 
- Stemming --> Reduce words to a root form. connecting, connected, connection --> connect- 
- Lemmatization --> Similar to stemming but smarter. running, runs, ran to run. This preserves actual dictionary meaning.- 
- Bag of Words --> count how many times each vocabulary word appears.- 
- Term Frequency (TF)- 
- Inverse Document Frequency (IDF)
- TF-IDF is feature engineering for text.
- TF-IDF asks two questions.
    - How important is this word inside this document?
    - Does this word appear in every document?
