```python
#read bronze
bronze_df = spark.table("dbw_agentic_ai_dev.telco_ai.bronze_telco")
display(bronze_df)
```

```python
#clean data
from pyspark.sql.functions import col, trim, when

silver_df = (bronze_df.withColumn("TotalCharges", when(trim(col("TotalCharges")) == "", None).otherwise(col("TotalCharges")))
    .withColumn("TotalCharges", col("TotalCharges").cast("double")))
```

```python
#verify schema
silver_df.printSchema()
```

```python
# check nulls
from pyspark.sql.functions import count, when
silver_df.select(count(when(col("TotalCharges").isNull(), True)).alias("null_totalcharges")).show()
```

```python
# write silver table
silver_df.write.format("delta").mode("overwrite").saveAsTable("dbw_agentic_ai_dev.telco_ai.silver_telco")
```
