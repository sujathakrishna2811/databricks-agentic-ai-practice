```python
# read silver
silver_df = spark.table("dbw_agentic_ai_dev.telco_ai.silver_telco")
```

```python
# convert churn column to binary
from pyspark.sql.functions import when, col
gold_df = (silver_df.withColumn("Churn_Flag", when(col("Churn") == "Yes", 1).otherwise(0)))
```

```python
# Add useful features
gold_df = gold_df.withColumn("Tenure_Group", when(col("tenure") < 12, "New").when(col("tenure") < 24, "Growing").otherwise("Loyal"))
```

```python
# create gold table
gold_df.write.format("delta").mode("overwrite").saveAsTable("dbw_agentic_ai_dev.telco_ai.gold_telco")
```

```python
%sql
-- validate gold table

SELECT
    Churn_Flag,
    COUNT(*) cnt
FROM dbw_agentic_ai_dev.telco_ai.gold_telco
GROUP BY Churn_Flag
```
