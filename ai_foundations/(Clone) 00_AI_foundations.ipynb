##### nlp_foundations

Traditional NLP (Natural Language Processing) Building Blocks

1. Text Preprocessing

   - Tokenization ->  Split into words. Sentence -> Words
   - Stop Words Removal -> Stop words are common words(the, a, is, of, and a, an, to)These usually add little meaning.
   - Stemming -> Reduce words to a root form. Ex: connecting, connected, connection --> connect
   - Lemmatization -> Similar to stemming but smarter. Ex: running, runs, ran to run. This preserves actual dictionary meaning.

2. Text Representation

   - Vocabulary ->  Collect every unique word.
   - Bag of Words -> count how many times each vocabulary word appears.
   - Term Frequency (TF)
   - Inverse Document Frequency (IDF)
   - TF-IDF

3. Feature Engineering

   - TF-IDF converts text into numerical features for Machine Learning.

4. Key Idea

   TF-IDF asks two questions:
   -  How important is this word in this document?
   -  Does this word appear in every document?

ML, NLP, RAG, Embeddings as four not unrelated topics but different ways of solving the same problem.

```text

Raw Information
        ↓
Represent it in a way the model understands
        ↓
Learn patterns or retrieve knowledge
        ↓
Produce a prediction or an answer

```

Before embeddings existed:

```text

Text
 ↓
TF-IDF
 ↓
Machine Learning

```
This worked surprisingly well for:

- Spam detection
- Sentiment analysis
- Ticket classification
- Email classification
- News categorization

Even today, TF-IDF is still used because it's:

- Fast
- Interpretable
- Doesn't require GPUs or LLMs
- Great baseline for text classification

complete traditional NLP pipeline is:

```text

Raw Text
      ↓
Text Cleaning
      ↓
Tokenization
      ↓
Vocabulary
      ↓
Bag of Words
      ↓
TF-IDF
      ↓
Feature Matrix
      ↓
Machine Learning Model
      ↓
Classification
      ↓
Confusion Matrix
Classification Report

```

The Evolution of NLP

```text

Sentence
    ↓
Tokenization
    ↓
Vocabulary
    ↓
Bag of Words
       ↓
Counts words
       ↓
TF-IDF
       ↓
Weights important words
       ↓
Embeddings
       ↓
Understand meaning
       ↓
Vector Search
       ↓
Retrieve semantically similar text
       ↓
RAG
       ↓
LLM generates answer

```

Notice how every stage solved a limitation of the previous one.

Pattern in the above every project follows the same high-level lifecycle:

```text

Data
      ↓
Preparation
      ↓
Representation
      ↓
AI Model
      ↓
Evaluation
      ↓
Deployment/Application

```

The only thing that changes is how the data is represented:

- Structured ML → engineered numeric features
- Traditional NLP → TF-IDF features
- Modern NLP → embeddings
- Computer Vision → image tensors
- Multimodal AI → combinations of text, images, and structured data

Core Principles

1. AI starts with data.
2. Every AI system converts raw data into a representation.

3. Better representations lead to better predictions.

4. Traditional ML uses engineered numeric features.

5. Traditional NLP uses TF-IDF features.

6. Modern NLP uses embeddings.

7. RAG retrieves knowledge before answering.

8. Agentic AI decides which tools to use before answering.

What do ML, NLP, RAG, Computer Vision, and Agentic AI have in common?

They all follow the same high-level lifecycle: data → representation → AI model → evaluation → deployment/application. The main difference is how the data is represented—engineered features, TF-IDF vectors, embeddings, image tensors, or multimodal representations.
