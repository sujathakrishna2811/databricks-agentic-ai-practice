
###### Traditional NLP Pipeline

```text

Raw Text
 ↓
Text Cleaning
 ↓
Tokenization
 ↓
Vocabulary
 ↓
Bag of Words
 ↓
TF-IDF
 ↓
Feature Matrix
 ↓
Machine Learning Model
 ↓
Classification

TF-IDF is feature engineering for text.
```


###### Before embeddings existed:

```text

Text
 ↓
TF-IDF
 ↓
Machine Learning

```
This worked surprisingly well for:

- Spam detection

Even today, TF-IDF is still used because it's:

- Fast
- Interpretable
- Doesn't require GPUs or LLMs
- Great baseline for text classification

What do ML, NLP, RAG, Computer Vision, and Agentic AI have in common?

Pattern for the below projects follows the same high-level lifecycle:
- ML
- NLP
- RAG
- Computer Vision
- Agentic AI

```text

Data
      ↓
Preparation
      ↓
Representation
      ↓
AI Model
      ↓
Evaluation
      ↓
Deployment/Application

```


The main difference is how the data is represented —engineered features

- Structured ML → engineered numeric features
- Traditional NLP → TF-IDF features
- Modern NLP → embeddings
- Computer Vision → image tensors
- Multimodal AI → combinations of text, images, and structured data

###### AI Core Principles

- AI starts with data.
- Every AI system converts raw data into a representation.
- Better representations lead to better predictions.
- Traditional ML uses engineered numeric features.
- Traditional NLP uses TF-IDF features.
- Modern NLP uses embeddings.
- RAG retrieves knowledge before answering.
- Agentic AI decides which tools to use before answering.

###### Big Learnings

- Databricks integrates data engineering, machine learning, and AI into a single unified platform.
- MLflow artifacts contain everything required to reproduce and deploy a model, including the model, signature, dependencies, and input examples.
- Traditional NLP using TF-IDF remains valuable because it is fast, interpretable, and an excellent baseline for text classification.
- Embeddings represent the meaning of text rather than simply counting words, enabling semantic search and retrieval.
- RAG does not train an LLM with new knowledge; instead, it retrieves relevant information and provides it as context before the LLM generates a response.
- Better retrieval leads to better AI responses—the quality of the retrieved context is often more important than the size of the language model.
- Agentic AI extends RAG by allowing an AI agent to reason, plan, and choose the appropriate tools before producing an answer.
- AI systems differ mainly in how data is represented: structured features for ML, TF-IDF vectors for traditional NLP, embeddings for modern NLP, and multimodal representations for vision and multimodal AI.
- Every AI application follows the same high-level lifecycle: Data → Representation → AI Model → Evaluation → Deployment/Application.
