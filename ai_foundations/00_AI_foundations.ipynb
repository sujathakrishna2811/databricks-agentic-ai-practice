
###### Traditional NLP Pipeline

### Chapter 3 - Modern NLP

%md
Modern NLP (Embeddings)

```text

Text
 ↓
Embeddings
 ↓
Vector
 ↓
LLM / Vector Search
 ↓
Answer
```


- What is an embedding?    
   - An embedding is a dense numerical vector that represents the semantic meaning of text.

- Why do we need embeddings? 
   - Embedding models convert text into vectors because Vector Search compares vectors, not raw text.

- Why are vectors better than TF-IDF?
  - TF-IDF matches important words, while embeddings capture the meaning of the entire sentence, allowing semantic search.

- Why do similar sentences have similar vectors?
  - Embedding models place semantically similar sentences close together in vector space.

- What is vector dimension? 
  - Vector dimension is the number of numerical values contained in an embedding vector.

- What is cosine similarity? 
  - Cosine similarity measures how similar two vectors are by comparing the angle between them rather than their length.

- Why do we store embeddings? 
  - Embeddings are stored in Delta tables so they can later be indexed by Vector Search and used for semantic retrieval, RAG, and Agentic AI applications.


###### The Evolution of NLP

```text

Sentence
    ↓
Tokenization
    ↓
Vocabulary
    ↓
Bag of Words
    ↓
Counts words
    ↓
TF-IDF
    ↓
Weights important words
    ↓
Embeddings
    ↓
Understand meaning
    ↓
Vector Search
    ↓
Retrieve semantically similar text
    ↓
RAG
    ↓
LLM generates answer

```

### AI Glossary

%md

##### Machine Learning

- Artificial Intelligence (AI) = 	A field of computer science that enables machines to perform tasks requiring human intelligence.

- Machine Learning (ML) = A subset of AI where models learn patterns from data instead of being explicitly programmed.

- Model = A trained algorithm that makes predictions or decisions from data.

- Feature = 	An input variable used by a machine learning model.

- Label / Target	= The value the model is trained to predict.

- Training Data = Data used to teach a machine learning model.

- Test Data = Data used to evaluate how well a trained model performs.

- Feature Engineering	 = Creating or transforming features to improve model performance.

- Classification	= Predicting discrete categories such as Spam or Not Spam.

- Regression = Predicting continuous numerical values.

- Accuracy = 	Percentage of predictions that are correct.

- Precision = Of predicted positives, how many were actually positive.

- Recall	=  Of actual positives, how many were correctly identified.

- F1 Score = 	Harmonic mean of Precision and Recall.

- F1 Macro = 	Average F1 score across all classes, treating each class equally.

- Confusion Matrix = 	A table showing correct and incorrect predictions by class.

- ROC-AUC	= Measures how well a binary classifier separates classes.

##### Traditional NLP

- NLP = Enables computers to understand and process human language.

- Tokenization = Splitting text into smaller units such as words.

- Stop Words =	Common words removed because they add little meaning.

- Stemming	= Reduces words to their root form by removing suffixes.

- Lemmatization	= Converts words to their dictionary root while preserving meaning.

- Vocabulary = Collection of all unique words in the dataset.

- Bag of Words = Represents text using word occurrence counts.

- TF  = Measures how often a word appears in one document.

- ID F = Measures how unique a word is across all documents.

- TF-IDF  = Converts text into weighted numerical features for machine learning.

- PyFunc Wrapper = A custom MLflow model wrapper that controls how input data is processed before calling the underlying model.

##### Databricks

- Databricks = Unified platform for data engineering, analytics, and AI.

- Unity Catalog = Centralized governance for data, models, and AI assets.

- Catalog = Top-level container for schemas and tables.

- Schema = Logical grouping of related database objects.

- Delta Table = Reliable table format supporting ACID transactions and versioning.

- Bronze Layer = Stores raw ingested data.

- Silver Layer = Stores cleaned and transformed data.

- Gold Layer = Stores business-ready curated data.

- Volume = Managed storage for files in Unity Catalog.

- Model Registry= Stores versioned machine learning models.

- Model Serving = Deploys models as REST API endpoints.

- Databricks Secret Scope = A secure, encrypted container used to store sensitive values such as API keys, passwords, PAT tokens, and connection strings. Secrets are retrieved in notebooks using dbutils.secrets.get() without exposing their values in code or notebook output.

##### Modern NLP / GenAI

- Embedding = Numerical vector representing the semantic meaning of text.

- Vector = Ordered list of numbers representing data mathematically.

- Semantic Search	= Finds information based on meaning instead of exact words.

- Vector Search = Retrieves similar embeddings using vector distance.

- LLM	= Large Language Model trained on massive text datasets.

- Prompt = Instructions given to an LLM.

##### RAG

- RAG	= Retrieves relevant knowledge before generating an answer.

- Retriever = Finds relevant documents for a query.

- Context	= Retrieved information passed to the LLM.

- Chunk = Small section of a document used for retrieval.

- Knowledge Base= Collection of documents used for retrieval.

##### Agentic AI

- Agent = AI system that plans, reasons, and performs tasks.

- Agentic AI	= AI that autonomously plans, selects tools, and executes multi-step workflows.

- Tool = External capability an agent can invoke, such as SQL or an API.

- Planning = Breaking a task into multiple steps before execution.

- Reasoning = Deciding what action to take next.

- Workflow = Sequence of actions executed to complete a task.

### Common Mistakes & Lessons Learned

Mistake

Thought model.pkl was the model registry.

Lesson

model.pkl is the serialized model.

Registry manages versions.

Mistake

Thought TF-IDF was a model.

Lesson

TF-IDF is only feature engineering.

Mistake

Confused artifact_path with registered model name.

Lesson

artifact_path identifies the logged model inside the MLflow run.

Mistake

Used Accuracy for multiclass comparison.

Lesson

F1 Macro is a better choice when all classes are equally important.

Mistake

Forgot input_example while logging.

Lesson

Serving uses it to infer the input schema.

Mistake:
Logged a raw sklearn TF-IDF pipeline and deployed it directly to Model Serving.

Lesson:
The pipeline was trained on a pandas Series, but Model Serving passed a pandas DataFrame based on the model signature. A PyFunc wrapper was needed to extract the text column before calling the pipeline.

###### MLflow Text Model Serving

###### Training

```text

DataFrame
      ↓
Select text column
      ↓
Series
      ↓
TF-IDF
      ↓
Classifier
```

###### Serving

```text
REST Request
      ↓
DataFrame
      ↓
Serving Model
      ↓
Prediction

```
###### Lesson
Always verify that the serving input type matches what the model expects during training.

Challenging deployment issue.

I deployed a text-classification model to Databricks Model Serving. The model performed correctly during local testing but returned the same class for every prediction through the endpoint. I systematically verified the model, artifacts, payload, endpoint configuration, and signature. I discovered the serving endpoint passed a DataFrame, while my TF-IDF pipeline expected a Pandas Series. I wrapped the pipeline in a custom MLflow PyFunc model that converted the DataFrame into the expected Series before prediction. After re-registering the model and deploying Version 2, the endpoint produced correct predictions.

The sequence to be followed to debug issue:

- Verified the model locally.
- Verified the registered model.
- Verified the endpoint configuration.
- Verified the model signature.
- Compared DataFrame vs Series behavior.
- Identified the input-type mismatch.
